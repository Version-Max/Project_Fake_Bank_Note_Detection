```python
import pandas as pd
import numpy as np
import lazypredict
#import xgboost as xgb
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
```

```python
df = pd.read_csv('BankNote_Authentication.csv')
```

```python
df.head()
```

```python
X = df.drop(['class'], axis = 1)
Y = df['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.25)
```

```python
classifier = LazyClassifier()
```

```python
models,predictions = classifier.fit(x_train, x_test, y_train, y_test)
```

```python
models
```

```python
# Looking from the above list of models I will be using RandomFOrest
```

```python
from sklearn.tree import DecisionTreeClassifier

model_DT = DecisionTreeClassifier()
model_DT.fit(x_train, y_train)
model_DT.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_RF = RandomForestClassifier()
model_RF.fit(x_train, y_train)
model_RF.score(x_test, y_test)
```

```python
# Considering that Random Forest might be overfitting
# I'll stick with Decision-Tree model
```
